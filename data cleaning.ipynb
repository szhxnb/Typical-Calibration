```python
folder_path = '/Users/hengxuzhu/Desktop/clear street/qr_takehome'
all_files = os.listdir(folder_path)
files = [file for file in all_files if file.endswith('.parquet')]
```

```python
dfs = [] 
for file in files: 
    file_path = os.path.join(folder_path, file) 
    df = pd.read_parquet(file_path) 
    df_new = dataClean(df, file[12:20]) 
    dfs.append(df_new) 
```

```python
combined_df = pd.concat(dfs, ignore_index=True).drop(columns=['sym', 'exch', 'Q1', 'Q2']).sort_values(by=['time']).set_index('time') 
```

```python
description = combined_df.describe()
description
```

```python
combined_df = combined_df.ffill() 
```

```python
combined_df.to_parquet('/Users/hengxuzhu/Desktop/clear street/QR project/Combined Clean Data.csv.parquet') 
```
